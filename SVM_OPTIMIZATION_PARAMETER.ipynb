```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.datasets import load_iris

# Load iris dataset
iris = load_iris()

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Define parameter ranges to search over
C_range = [0.1, 1, 10, 100]
gamma_range = [0.1, 1, 10, 100]
kernel_range = ['linear', 'rbf']

# Initialize variables to store scores and best parameters
scores = []
params = []

# Loop over parameter ranges to find best parameters
for C in C_range:
    for gamma in gamma_range:
        for kernel in kernel_range:
            
            # Create SVM model with current parameters
            svm = SVC(C=C, gamma=gamma, kernel=kernel)

            # Train and evaluate model on training set
            history = svm.fit(X_train, y_train)
            score = svm.score(X_test, y_test)
            
            # Store score and parameters
            scores.append(score)
            params.append({'C': C, 'gamma': gamma, 'kernel': kernel})
            
# Plot graph of accuracy for different parameters and iterations
fig, ax = plt.subplots()
ax.plot(range(len(scores)), scores, 'bo')
ax.set_xlabel('Iterations')
ax.set_ylabel('Accuracy')
ax.set_title('SVM parameter optimization')
ax.set_xticks(range(len(scores)))
ax.set_xticklabels(params, rotation=90)
plt.show()
```
